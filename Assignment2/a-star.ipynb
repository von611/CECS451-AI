```python
from_city = input("From city: ")
to_city = input("To city: ")
min_city = from_city
route = []
route.append(from_city)

with open('map.txt') as file:
    cities = list(file.read().splitlines()) #parse the map file into lines and store each one in a list
    
with open('distances.txt') as file:
    distances = list(file.read().splitlines()) #parse the file into lines and store each one as in a list    

min_distance = 1000
total_distance = 0
while from_city != to_city:
    for city in cities:
        city_neighbors = city.split('-')
        if city_neighbors[0] == from_city:
            distance = 0
            city_neighbors[1] = city_neighbors[1].replace('(',' ')
            city_neighbors[1] = city_neighbors[1].replace(')','')
            neighbors = city_neighbors[1].split(',')
            for neighbor in neighbors:
                neighbor_distance = neighbor.split(' ')
                for d in distances:
                    city_sld = d.split('-')
                    if city_sld[0] == neighbor_distance[0]:
                        if min_distance >= int(neighbor_distance[1]) + int(city_sld[1]):
                            min_distance = int(neighbor_distance[1]) + int(city_sld[1])
                            distance = int(neighbor_distance[1])
                            min_city = neighbor_distance[0]
                        break
            total_distance += distance
            from_city = min_city
            route.append(min_city)

hyphen = '-'            
best_route = hyphen.join(route)
print('Best Route: ' + best_route)
print('Total Distance: ' + str(total_distance))
```
