```python
import sys
import time
from board import Board
            
if __name__ == '__main__':
    test = Board(5)
    test.fitness()
    localCount=0
    restart=0
    boardFit=test.get_fit()
    startTime = time.time()

    while test.get_fit() !=0:
        
        for i in range(test.n_queen):
            minIndex=0
            currFit=sys.maxsize
            for j in range(test.n_queen):

                test.zeroRow(i)
                test.flip(i,j)
                test.fitness()
                if test.get_fit()<currFit:
                    currFit=test.get_fit()
                    minIndex=j
            test.zeroRow(i)
            test.flip(i,minIndex)

        if boardFit == test.get_fit():
            localCount+=1
            
        else:
            localCount=0
        if localCount >= 3: # if result is the same 3 times in a row then reset the board
            test=Board(5)
            test.fitness()
            restart+=1
        boardFit=test.get_fit()
        
    endTime=int((time.time()-startTime)*1000)
    print('Running Time:',endTime,'ms')
    print('Number of restart:',restart)
    test.show()
```
